### 1) String Concatenation (Join Karna)

```python
first = "AI"
second = "Engineer"
print(type(first))
print(first + " " + second)
```

### 2) String Repetition

```python
print("Hire " * 3)
```

### 3) Indexing (Character Access Karna)

```python
word = "Python"
print(word[4])
```

```python
print(word[-2])
```

### 4) Slicing (Part Extract Karna)

```python
word = "Artificial Intelligence"

print(word[0:10])
# means 5 wala index include nahe ho ga
```

### 5) Length of String

```python
text = "Machine "
print(len(text))
```

### 6) String Methods

```python
text = "python  automation"
print(text.upper())
print(text.lower())
print(text.strip())
```

```python
# Strip (Spaces Remove)
text = "   AI   "
print(text.strip())
# Strip (Spaces Remove)
text = "Hello AI"
print(text.replace("AI", "World"))
# Replace
text = "Hello AI"
print(text.replace("AI", "World"))
# Split
text = "AI is powerful"
print(text.split())
# Find
text = "Artificial Intelligence"
print(text.find("Intelligence"))
```

### 7) String Formatting

```python
# Old Style
name = "Hira"
age = 20
# print(name + age) Error
print("My name is", name, "and I am", age)
```

```python
# f-String (Best Method)
print(f"My name is {name} and I am {age}")
```

### 8) Checking in String

```python
text = "AI Engineer"

print("AI" in text)
```

### 9) Escape Characters

```python
print("Hello\nWorld")
print("Hello\tWorld")
print("Hello\"World")
print("HelloWorld")
```
