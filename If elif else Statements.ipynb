# *1. If, elif, else Statements*

```python
if condition:
    code
elif condition:
    code
else:
    code
```

```python
age = 18

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

### Multiple Conditions

```python
marks = 75

if marks >= 90:
    print("A")
elif marks >= 70:
    print("B")
elif marks >= 50:
    print("C")
else:
    print("Fail")
```

# Nested Conditionals

```python
age = 20
citizen = True

if age >= 18:
    if citizen:
        print("Eligible to vote")
    else:
        print("Not a citizen")
else:
    print("Too young")
```

# Conditional (Ternary) Expression

```python
age = 17


print("Adult" if age >= 18 else "Minor")
```

```python
num = 7

print("Even" if num % 2 == 0 else "Odd")
```
