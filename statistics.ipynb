```python
# ============================================================
# 🔹 1. WHAT IS STATISTICS
# ============================================================

# Definition (Aligned with your document):
# Statistics is a branch of mathematics that involves collecting,
# analysing, interpreting, and presenting data.

# It provides tools and methods to understand large amounts of data
# and to draw conclusions and make decisions based on the data.

# Applications (from document):
# - Business → customer behavior, demand forecasting
# - Medical → clinical trials, disease risk analysis
# - Government → surveys, polling
# - Environmental → climate research

# Why we need it:
# - To convert raw data into useful information
# - To support decision making

# Disadvantages:
# - Depends on data quality
# - Can lead to wrong conclusions if misused

data = [10, 20, 30, 40]
print("Mean:", sum(data)/len(data))
```

```python
# ============================================================
# 🔹 2. TYPES OF STATISTICS
# ============================================================

# Definition:
# Descriptive Statistics:
# Deals with the collection, organization, analysis,
# interpretation, and presentation of data.
# It summarizes the main features of a dataset
# WITHOUT making predictions.

# Inferential Statistics:
# Deals with making conclusions and predictions about a population
# based on a sample.
# Uses probability, hypothesis testing, and regression.

# Why needed:
# - Descriptive → understand data
# - Inferential → predict unknown population values

# Disadvantages:
# - Inferential results depend on sample quality

import numpy as np

data = np.array([10, 20, 30, 40])
print("Descriptive Mean:", np.mean(data))
print(type(data))

sample = np.random.choice(data, 2)
print("Sample Mean:", np.mean(sample))
```

```python
# ============================================================
# 🔹 3. POPULATION VS SAMPLE
# ============================================================

# Definition:
# Population:
# The entire group of individuals or objects that we are interested in studying.
# It is the complete set of observations.

# Sample:
# A subset of the population selected for study.
# Used to estimate population characteristics.

# Important Points (from document):
# - Sample Size
# - Random Selection
# - Representative Sample

# Why needed:
# - Population is often too large or unavailable

# Disadvantages:
# - Poor sampling leads to wrong conclusions

population = [10, 20, 30, 40, 50]
sample = [20, 40]

print("Population Mean:", sum(population)/len(population))
print("Sample Mean:", sum(sample)/len(sample))
```

```python
# ============================================================
# 🔹 4. PARAMETER VS STATISTIC
# ============================================================

# Definition:
# Parameter:
# A characteristic of a population (usually unknown).

# Statistic:
# A characteristic of a sample used to estimate the parameter.

# Goal:
# Use sample statistics to make inferences about population parameters.

import numpy as np

population = np.array([10, 20, 30, 40])
sample = np.array([20, 30])

print("Parameter:", np.mean(population))
print("Statistic:", np.mean(sample))
```
