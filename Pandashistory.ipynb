What is a Series?
A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.

Labels
If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.

This label can be used to access a specified value.
With the index argument, you can name your own labels.


Key/Value Objects as Series
You can also use a key/value object, like a dictionary, when creating a Series.
Note: The keys of the dictionary become the labels.

```python
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar[0]) # you can access rows and columns using labels
indexargument= pd.Series(a,index=["x","y","z"]) # you can also specify index
print(indexargument)
print(indexargument['z']) # you can access values with index
print(myvar)

# using key value pair like dictionary
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"]) #To select only some of the items in the dictionary, use the index argument and specify only the items you want to include in the Series.
print(myvar["day1"])


print(myvar)
```

DataFrames
Data sets in Pandas are usually multi-dimensional tables, called DataFrames.

Series is like a column, a DataFrame is the whole table.

Locate Row
Pandas use the loc attribute to return one or more specified row(s)


```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

#refer to the row index:
print(df.loc[0]) #Note: This example returns a Pandas Series.

#use a list of indexes:
print(df.loc[[0, 1]]) #Note: When using [], the result is a Pandas DataFrame.

df2 = pd.DataFrame(data, index = ["day1", "day2", "day3"]) # name index like in series 
print(df2)
#refer to the named index:
print(df2.loc["day2"])


print(df) 
```

Read CSV Files
A simple way to store big data sets is to use CSV files (comma separated files).

CSV files contains plain text and is a well know format that can be read by everyone including Pandas



```python
import pandas as pd
df = pd.read_csv("data.csv")
# df.head() # return only first five rows if you not pass any arguments
# # df.tail() # return only last five rows if you not pass any arguments
# print(df) # if you use to display data its only show first and last five dataset 
# print(df.to_string()) # if you want to show full data set you can use to_string() method
df[['Brand']] # show column Name with bold 
#df['Brand'] # selecting and indexing selecting single column \

df[['Brand','Sale_ID']] # selecting mulitple columns
#df['Brand','Sale_ID'] # its give error you can not access muliple column using this syntax
df.iloc[0] # return 1 row of the data set 
df.iloc[0:1] # return 1 row of the data set with bold keyword 
df.loc[0,'Brand'] # row 0 , column Name
df.iloc[0,2] # row 0 , column index 2
df[df['Price_USD'] > 80] # apply condition 
df[(df['Price_USD'] > 80 ) & (df["Country"] == "UK")] # apply condition 
```

```python
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "Ahmed", "Ali", None],
    "Marks": [85, 90, 75, 85, None],
    "City": ["Lahore", "Karachi", "Islamabad", None, "Karachi"]
}

df = pd.DataFrame(data)
print(df["Name"].unique())
print(df["Name"].value_counts())
print(df.columns)
print(df.index)
print(df)
```

```python
df.sort_values('Marks') # sort values according to Marks 
df.sort_values('Marks', ascending=False) # descending order
df.sort_values(['City','Marks'])
df.isnull() # return false if the value is not null other wise return true
df.isnull().sum() # return how many is null in the columns 
```

```python
df.dropna() # remove None row from the data set 
df.fillna(10,inplace=True) # replace None values with specific value that you provide , and inplace change original data set 
print(df)
```

```python
df['City'].replace('Lahore','LHR') # replace one value
df.replace({'Lahore':'LHR' , 'Karachi': 'KHI'}) # replace more than one value 
```

```python
df.drop('City',axis=1) # Drop column when you drop column also specified the axis 1 which refer to the column 
df.drop(0) # drop row with zero index
```
