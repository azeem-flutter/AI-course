What is a Series?
A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.

Labels
If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.

This label can be used to access a specified value.
With the index argument, you can name your own labels.


Key/Value Objects as Series
You can also use a key/value object, like a dictionary, when creating a Series.
Note: The keys of the dictionary become the labels.

```python
# ============================================================
# HANDLING MISSING VALUES IN PANDAS - COMPLETE BEGINNER GUIDE
# ============================================================

# Import required libraries
import pandas as pd
import numpy as np

# ------------------------------------------------------------
# 1) WHAT ARE MISSING VALUES?
# ------------------------------------------------------------
# Missing values in pandas appear as:
# - NaN  (Not a Number) -> for numerical data
# - None -> for object/categorical data
# - NaT  (Not a Time)    -> for datetime data

# Creating a sample dataset
data = {
    "Age": [20, 25, np.nan, 30],  # numerical column with missing value
    "Salary": [20000, 25000, 1000000, None],  # contains outlier and missing value
    "City": ["FSD", "LHR", None, "ISB"],  # categorical column
    "Date": pd.to_datetime(["2024-01-01", None, "2024-01-03", "2024-01-04"])
}

df = pd.DataFrame(data)

print("Original DataFrame:\n", df)

# ------------------------------------------------------------
# 2) DETECTING MISSING VALUES
# ------------------------------------------------------------

# Check True/False for missing values
print("\nMissing values (True means missing):\n", df.isnull())

# Count missing values per column
print("\nCount of missing values:\n", df.isnull().sum())

# Why we do this?
# Because you NEVER fix data blindly.
# First you measure how much is missing.


# ------------------------------------------------------------
# 3) UNDERSTANDING MEAN, MEDIAN, MODE
# ------------------------------------------------------------

# ---------------------
# MEAN
# ---------------------
# Formula:
# mean = (sum of all values) / number of values
#
# Used for:
# - Numerical data
# - Normally distributed data
#
# NOT good for:
# - Data with outliers
# - Skewed data

print("\nMean of Salary (ignores NaN):", df["Salary"].mean())

# ---------------------
# MEDIAN
# ---------------------
# Median = middle value after sorting
#
# If odd count → middle value
# If even count → average of two middle values
#
# Used for:
# - Skewed data
# - Data with outliers

print("Median of Salary:", df["Salary"].median())

# ---------------------
# MODE
# ---------------------
# Mode = most frequent value
#
# Used for:
# - Categorical data
# - Repeated values

print("Mode of City:", df["City"].mode()[0])


# ------------------------------------------------------------
# 4) WHAT IS AN OUTLIER?
# ------------------------------------------------------------

# Outlier = extremely large or small value compared to others
#
# Example:
# 20000, 25000, 27000, 1000000
#
# 1000000 is an outlier because it is far away from others
#
# Why dangerous?
# Because it distorts the mean.

# Compare mean vs median
print("\nSalary Mean:", df["Salary"].mean())
print("Salary Median:", df["Salary"].median())

# If mean >> median → data likely right-skewed


# ------------------------------------------------------------
# 5) WHAT IS SKEW?
# ------------------------------------------------------------

# Skew tells whether data is symmetric or not.
#
# If skew ≈ 0 → symmetric (normal distribution)
# If skew > 0 → right-skewed (long tail to right)
# If skew < 0 → left-skewed (long tail to left)

print("\nSalary Skewness:", df["Salary"].skew())

# Right skew example:
# Most values small, few extremely large values
#
# Left skew example:
# Most values high, few extremely small values


# ------------------------------------------------------------
# 6) METHODS TO HANDLE MISSING VALUES
# ------------------------------------------------------------

# ============================
# METHOD 1: DROP ROWS
# ============================

# Removes rows with ANY missing value
df_drop_rows = df.dropna()

# Use when:
# - Dataset is large
# - Missing values are very few (<5%)

print("\nAfter dropping rows:\n", df_drop_rows)


# ============================
# METHOD 2: DROP COLUMNS
# ============================

# Removes entire column with missing values
df_drop_cols = df.dropna(axis=1)

# Use when:
# - Column is mostly empty
# - Column is not important

print("\nAfter dropping columns:\n", df_drop_cols)


# ============================
# METHOD 3: FILL WITH MEAN
# ============================

df_mean_fill = df.copy()
df_mean_fill["Age"] = df_mean_fill["Age"].fillna(df_mean_fill["Age"].mean())

# Use when:
# - Data is numerical
# - No extreme outliers
# - Distribution roughly symmetric

print("\nAfter filling Age with Mean:\n", df_mean_fill)


# ============================
# METHOD 4: FILL WITH MEDIAN
# ============================

df_median_fill = df.copy()
df_median_fill["Salary"] = df_median_fill["Salary"].fillna(df_median_fill["Salary"].median())

# Use when:
# - Data is numerical
# - Contains outliers
# - Right or left skewed

print("\nAfter filling Salary with Median:\n", df_median_fill)


# ============================
# METHOD 5: FILL WITH MODE
# ============================

df_mode_fill = df.copy()
df_mode_fill["City"] = df_mode_fill["City"].fillna(df_mode_fill["City"].mode()[0])

# Use when:
# - Data is categorical
# - You want most common category

print("\nAfter filling City with Mode:\n", df_mode_fill)


# ============================
# METHOD 6: FILL WITH CONSTANT VALUE
# ============================

df_constant_fill = df.copy()
df_constant_fill["Salary"] = df_constant_fill["Salary"].fillna(0)

# Use when:
# - Missing value itself has meaning
# Example:
# Salary = 0 means unemployed

print("\nAfter filling Salary with 0:\n", df_constant_fill)


# ============================
# METHOD 7: FORWARD FILL (Time Series)
# ============================

# Time Series Data:
# Data ordered by time (stock price, temperature, sales per day)

df_time_fill = df.copy()
df_time_fill = df_time_fill.sort_values("Date")

df_time_fill["Age"] = df_time_fill["Age"].fillna(method="ffill")

# Forward fill copies previous value downward
#
# Use when:
# - Data is sequential
# - Order matters
# - Time-based data

print("\nAfter Forward Fill:\n", df_time_fill)


# ------------------------------------------------------------
# 5️⃣ FORWARD FILL & BACKWARD FILL
# ------------------------------------------------------------

# Forward Fill (ffill)
# Copies previous valid value downward

df["Age"].fillna(method="ffill", inplace=True)

# Backward Fill (bfill)
# Copies next valid value upward

df["Age"].fillna(method="bfill", inplace=True)

# IMPORTANT:
# ffill cannot fill FIRST value if it is NaN
# bfill cannot fill LAST value if it is NaN

# Example:
# NaN, NaN, 10, 20, NaN
# After ffill → NaN, NaN, 10, 20, 20
# After bfill → 10, 10, 10, 20, NaN
# After ffill + bfill → 10, 10, 10, 20, 20


# WHY USE ffill/bfill:
# Only for Time Series Data


# ============================
# METHOD 8: BACKWARD FILL
# ============================

df_backward_fill = df.copy()
df_backward_fill["Age"] = df_backward_fill["Age"].fillna(method="bfill")

# Backward fill copies next value upward
#
# Also used in time series data

print("\nAfter Backward Fill:\n", df_backward_fill)


# ------------------------------------------------------------
# 7) PROFESSIONAL DECISION RULES
# ------------------------------------------------------------

# 1) If numerical + normal distribution → use MEAN
# 2) If numerical + skewed/outliers → use MEDIAN
# 3) If categorical → use MODE
# 4) If time series → use forward/backward fill
# 5) If too many missing values → consider dropping column
# 6) Always check skew and outliers before deciding

print("\n--- END OF COMPLETE GUIDE ---")
```

```python
# ===============================================================
# PANDAS CONCEPTS: GROUPBY, DISCRETIZATION, AND BINNING
# ===============================================================

# ---------------------------------------------------------------
# WHAT IS DISCRETIZATION?
# ---------------------------------------------------------------
# Discretization means converting continuous numerical values
# into categories (groups).
#
# Example:
# Age values: 21, 25, 30, 35, 40
#
# Instead of using exact numbers, we create categories:
#
# 20–30  -> Young
# 30–40  -> Adult
# 40–50  -> Senior
#
# So numbers become categories.
#
# The most common way to perform discretization in pandas
# is by using BINNING.
#
# ---------------------------------------------------------------
# WHAT IS BINNING?
# ---------------------------------------------------------------
# Binning means dividing numerical data into ranges (called bins).
#
# Example bins:
# [20–30], [30–40], [40–50]
#
# Each value will fall inside one of these bins.
#
# In pandas we use:
# pd.cut()  -> equal width bins
# pd.qcut() -> equal frequency bins
#
# ---------------------------------------------------------------
# WHAT IS GROUPBY?
# ---------------------------------------------------------------
# GroupBy means splitting data into groups based on a column
# and then performing calculations on each group.
#
# Example:
# If we group by Department, pandas will separate rows like:
#
# IT group
# HR group
# Finance group
#
# Then we can calculate:
# - mean()
# - sum()
# - count()
# - max()
# - min()
#
# This process is called:
# Split → Apply → Combine
#
# ===============================================================
# NOW WE IMPLEMENT THESE CONCEPTS USING CODE
# ===============================================================

import pandas as pd  # import pandas library

# ---------------------------------------------------------------
# STEP 1: CREATE A DATASET
# ---------------------------------------------------------------
# Here we create a simple dataset containing employee information

data = {
    "Name": ["Ali", "Sara", "Ahmed", "Hamza", "Ayesha", "Zain"],
    "Department": ["IT", "IT", "HR", "HR", "Finance", "Finance"],
    "Age": [21, 25, 30, 35, 40, 45],
    "Salary": [4000, 5000, 3500, 3800, 6000, 6200]
}

# Convert dictionary into pandas DataFrame
df = pd.DataFrame(data)

# Print the original dataset
print("ORIGINAL DATASET")
print(df)


# ---------------------------------------------------------------
# STEP 2: DISCRETIZATION USING BINNING (pd.cut)
# ---------------------------------------------------------------
# We convert Age numbers into Age categories.

# Define bin ranges
# These numbers represent boundaries of the ranges
bins = [20, 30, 40, 50]

# Define labels for each range
labels = ["Young", "Adult", "Senior"]

# pd.cut() places each Age value into one of these bins
df["Age_Group"] = pd.cut(df["Age"], bins=bins, labels=labels)

# Now the dataset contains a new column called Age_Group
print("\nDATASET AFTER BINNING (DISCRETIZATION)")
print(df)


# ---------------------------------------------------------------
# HOW pd.cut() WORKS
# ---------------------------------------------------------------
# bins = [20, 30, 40, 50]
#
# This creates ranges:
#
# (20–30] -> Young
# (30–40] -> Adult
# (40–50] -> Senior
#
# Example mapping:
#
# Age 21 -> Young
# Age 25 -> Young
# Age 30 -> Adult
# Age 35 -> Adult
# Age 40 -> Senior
# Age 45 -> Senior


# ---------------------------------------------------------------
# STEP 3: GROUPBY OPERATION
# ---------------------------------------------------------------
# Now we use GroupBy to analyze data.
#
# Here we group rows based on Age_Group
# and calculate the average salary for each group.

salary_by_age = df.groupby("Age_Group")["Salary"].mean()

print("\nAVERAGE SALARY PER AGE GROUP")
print(salary_by_age)


# ---------------------------------------------------------------
# WHAT GROUPBY DID INTERNALLY
# ---------------------------------------------------------------
# pandas performed three steps:
#
# 1) SPLIT
# Young group
# Adult group
# Senior group
#
# 2) APPLY
# mean() applied to Salary column
#
# 3) COMBINE
# Results merged into final output.


# ---------------------------------------------------------------
# STEP 4: MULTIPLE AGGREGATIONS WITH GROUPBY
# ---------------------------------------------------------------
# Instead of one calculation, we can compute many statistics.

stats = df.groupby("Department")["Salary"].agg(["mean", "sum", "max", "min", "count"])

print("\nDEPARTMENT SALARY STATISTICS")
print(stats)


# ---------------------------------------------------------------
# STEP 5: BINNING USING pd.qcut()
# ---------------------------------------------------------------
# pd.qcut() creates bins with equal number of data points.
#
# If q=3, pandas divides data into 3 groups with equal size.

df["Age_Group_qcut"] = pd.qcut(df["Age"], q=3, labels=["Young", "Adult", "Senior"])

print("\nDATASET USING qcut (EQUAL FREQUENCY BINNING)")
print(df)


# ---------------------------------------------------------------
# DIFFERENCE BETWEEN cut() AND qcut()
# ---------------------------------------------------------------
#
# pd.cut()
# - creates bins with equal width
# - example ranges: 20–30, 30–40, 40–50
#
# pd.qcut()
# - creates bins with equal number of rows
# - each bin contains same number of data points
#
# ---------------------------------------------------------------
# FINAL DATA ANALYSIS WORKFLOW
# ---------------------------------------------------------------
#
# Raw Data
#     ↓
# Discretization (pd.cut or pd.qcut)
#     ↓
# Create categories
#     ↓
# GroupBy
#     ↓
# Aggregation (mean, sum, count, etc.)
#
# ===============================================================
```
