# *While, for loops and use of enumerate*

# While Loop

```python
ðŸ‘‰ Jab tak condition True ho, loop chalta rehta hai
```

```python
while condition:
    code
```

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

# For Loop

```python
ðŸ‘‰ Sequence par iterate karta hai (list, string, range, etc.)
```

```python
for i in range(5):
    print(i)
```

```python
fruits = ["apple", "banana", "mango"]

for f in fruits:
    print(f)
```

# enumerate() ka Use

```python
ðŸ‘‰ Index + value dono deta hai ðŸ”¥
```

```python
fruits = ["apple", "banana", "mango"]

for index, value in enumerate(fruits):
    print(index, value)
```

# Nested Loops

```python
for i in range(3):
    for j in range(2):
        print(i, j)
```

# List Comprehension

```python
squares = [x*x for x in range(5)]
print(squares)
```

```python
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

# Iterables

```python
for ch in "HELLO":
    print(ch)
```

# Iterators

```python
nums = [10, 20, 30]

it = iter(nums)

print(next(it))
print(next(it))
print(next(it))
```
