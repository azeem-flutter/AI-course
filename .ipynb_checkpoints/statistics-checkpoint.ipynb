```python
# ============================================================
# 🔹 1. WHAT IS STATISTICS
# ============================================================

# Definition (Aligned with your document):
# Statistics is a branch of mathematics that involves collecting,
# analysing, interpreting, and presenting data.

# It provides tools and methods to understand large amounts of data
# and to draw conclusions and make decisions based on the data.

# Applications (from document):
# - Business → customer behavior, demand forecasting
# - Medical → clinical trials, disease risk analysis
# - Government → surveys, polling
# - Environmental → climate research

# Why we need it:
# - To convert raw data into useful information
# - To support decision making

# Disadvantages:
# - Depends on data quality
# - Can lead to wrong conclusions if misused

data = [10, 20, 30, 40]
print("Mean:", sum(data)/len(data))
```

```python
# ============================================================
# 🔹 2. TYPES OF STATISTICS
# ============================================================

# Definition:
# Descriptive Statistics:
# Deals with the collection, organization, analysis,
# interpretation, and presentation of data.
# It summarizes the main features of a dataset
# WITHOUT making predictions.

# Inferential Statistics:
# Deals with making conclusions and predictions about a population
# based on a sample.
# Uses probability, hypothesis testing, and regression.

# Why needed:
# - Descriptive → understand data
# - Inferential → predict unknown population values

# Disadvantages:
# - Inferential results depend on sample quality

import numpy as np

data = np.array([10, 20, 30, 40])
print("Descriptive Mean:", np.mean(data))
print(type(data))

sample = np.random.choice(data, 2)
print("Sample Mean:", np.mean(sample))
```

```python
# ============================================================
# 🔹 3. POPULATION VS SAMPLE
# ============================================================

# Definition:
# Population:
# The entire group of individuals or objects that we are interested in studying.
# It is the complete set of observations.

# Sample:
# A subset of the population selected for study.
# Used to estimate population characteristics.

# Important Points (from document):
# - Sample Size
# - Random Selection
# - Representative Sample

# Why needed:
# - Population is often too large or unavailable

# Disadvantages:
# - Poor sampling leads to wrong conclusions

population = [10, 20, 30, 40, 50]
sample = [20, 40]

print("Population Mean:", sum(population)/len(population))
print("Sample Mean:", sum(sample)/len(sample))
```

```python
# ============================================================
# 🔹 4. PARAMETER VS STATISTIC
# ============================================================

# Definition:
# Parameter:
# A characteristic of a population (usually unknown).

# Statistic:
# A characteristic of a sample used to estimate the parameter.

# Goal:
# Use sample statistics to make inferences about population parameters.

import numpy as np

population = np.array([10, 20, 30, 40])
sample = np.array([20, 30])

print("Parameter:", np.mean(population))
print("Statistic:", np.mean(sample))
```

```python
# ============================================================
# 🔹 5. INFERENTIAL STATISTICS (TOPICS)
# ============================================================

# Definition:
# Inferential statistics involves using sample data to make
# predictions or inferences about a population.

# Main Techniques (from document):

# 1. Hypothesis Testing:
# Testing a claim about a population

# 2. Confidence Interval:
# Estimating a range for population parameter

# 3. ANOVA:
# Comparing means across multiple groups

# 4. Regression:
# Relationship between variables

# 5. Chi-Square Test:
# Association between categorical variables

# 6. Sampling Techniques:
# Ensuring sample is representative

# 7. Bayesian Statistics:
# Updating probability based on new data

# Disadvantages:
# - Results depend heavily on sample quality
# - Requires assumptions
```

```python
# ============================================================
# 🔹 6. MEASURE OF CENTRAL TENDENCY
# ============================================================

# Definition:
# A measure of central tendency is a statistical measure that represents
# a typical or central value of a dataset. It summarizes the dataset
# by providing a single value that best describes the entire data.

# Why we need it:
# - To understand the overall behavior of data
# - To simplify large datasets into one representative value

# Limitations:
# - Does not describe spread of data
# - Different measures give different results depending on data


# ============================================================
# 🔸 MEAN (Average)
# ============================================================

# Definition:
# Mean is the sum of all observations divided by the total number of observations.

# Advantages:
# - Uses all values in dataset
# - Easy to calculate

# Disadvantages:
# - Highly affected by outliers (extreme values)

data = [10, 20, 30, 40]

# Simple Python (BEST for beginners)
mean = sum(data) / len(data)
print("Mean:", mean)

# Using NumPy (WHY? → faster + used in ML)
import numpy as np
print("Mean using NumPy:", np.mean(data))


# ============================================================
# 🔸 MEDIAN
# ============================================================

# Definition:
# Median is the middle value when the data is arranged in ascending or descending order.

# Advantages:
# - Not affected by outliers
# - Better for skewed data

# Disadvantages:
# - Does not use all values
# - Needs sorting

data = [10, 20, 30, 40, 100]

sorted_data = sorted(data)
median = sorted_data[len(data)//2]
print("Median:", median)

# NumPy version (faster for large datasets)
print("Median using NumPy:", np.median(data))


# ============================================================
# 🔸 MODE
# ============================================================

# Definition:
# Mode is the value that appears most frequently in a dataset.

# Advantages:
# - Useful for categorical data
# - Easy to understand

# Disadvantages:
# - Can have multiple modes
# - Sometimes no mode exists

from collections import Counter

data = [1, 2, 2, 3, 3]

freq = Counter(data)
max_freq = max(freq.values())

mode = [k for k,v in freq.items() if v == max_freq]
print("Mode:", mode)


# ============================================================
# 🔸 WEIGHTED MEAN
# ============================================================

# Definition:
# Weighted mean is calculated by multiplying each value by its weight,
# summing them, and dividing by total weights.

# Why needed:
# - When some values are more important than others

# Example:
# GPA calculation, marks with different credit hours

# Disadvantages:
# - Wrong weights → wrong result

values = [10, 20, 30]
weights = [1, 2, 3]

weighted_mean = sum(v*w for v, w in zip(values, weights)) / sum(weights)
print("Weighted Mean:", weighted_mean)


# ============================================================
# 🔸 TRIMMED MEAN
# ============================================================

# Definition:
# Trimmed mean is calculated by removing a fixed percentage of the smallest
# and largest values and then computing the mean of remaining data.

# Why needed:
# - Reduces effect of outliers

# Disadvantages:
# - Loss of data
# - Choice of trimming percentage can affect result

data = [10, 20, 30, 40, 100]

# Step 1: sort data
sorted_data = sorted(data)

# Step 2: remove extreme values (here removing 1 smallest and 1 largest)
trimmed_data = sorted_data[1:-1]

# Step 3: calculate mean
trimmed_mean = sum(trimmed_data) / len(trimmed_data)

print("Trimmed Mean:", trimmed_mean)


# ============================================================
# 🔹 IMPORTANT: WHEN TO USE WHICH?
# ============================================================

# Mean → when data has no outliers
# Median → when data has outliers
# Mode → when dealing with categories
# Weighted Mean → when importance differs
# Trimmed Mean → when data has extreme values
```
